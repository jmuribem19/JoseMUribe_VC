```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
im=cv.imread('/content/drive/MyDrive/EAFIT/2024-2/VisioÃÅn por computadora/cat (1).jpg')
green=im[:,:,1]
plt.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv.cvtColor(green, cv.COLOR_BGR2RGB))
plt.show()
```

```python
escaled = green/255
```

```python
vert = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])
vert
```

```python
hor = np.array([[1,0,-1],
                     [2,0,-2],
                     [1,0,-1]])
hor
```

```python
imVer=cv.filter2D(escaled,-1,vert)
plt.imshow(imVer, cmap="gray")
plt.show()
```

```python
imHor=cv.filter2D(escaled,-1,hor)
plt.imshow(imHor,cmap="gray")
plt.show()
```

```python
sobel = np.sqrt(imHor**2 + imVer**2)
plt.imshow(sobel, cmap="gray")
plt.show()
```
